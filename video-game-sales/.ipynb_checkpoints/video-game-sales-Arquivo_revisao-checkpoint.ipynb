```python
# Análise de dados Sobre o arquivo Video game Sales
# https://www.kaggle.com/code/davisxergs/video-game-sales-analysis
```

#### This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

#### Fields include

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

###### The script to scrape the data is available at https://github.com/GregorUT/vgchartzScrape.
###### It is based on BeautifulSoup using Python.
###### There are 16,598 records. 2 records were dropped due to incomplete information.

```python
# Libries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('vgsales.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

<h1><b> Limpando, transformando e Discretizando osdados </b></h1>

# Tratando dados ausentes

```python
df.isnull().sum()
```

<h1> Removendo linhas ausentes </h1>

```python
shape_before = df.shape[0]
shape_curr = df.dropna().shape[0]
print(f"Linhas removidas {shape_before - shape_curr} equivale a {round((shape_before - shape_curr)/df.shape[0]*100,5)}%")
```

```python
df.dropna(inplace = True)
df.shape[0]
```

```python
df = df.append(pd.Series(), ignore_index=True)
df.tail()
```

<h4>Removendo linhas que contém somente NaN</h4>

```python
shape_before = df.shape[0]
shape_curr = df.dropna(how = 'all').shape[0]
print(f"Linhas removidas {shape_before - shape_curr} equivale a {round((shape_before - shape_curr)/df.shape[0]*100,5)}%")
```

```python
df.dropna(how ='all', inplace = True)
df.shape[0]
```

```python
df.tail()
```

<h3> Removendo a coluna Rank </h3>

```python
df.drop('Rank', axis = 1, inplace = True)
df.head(5)
```

### Removendo caso tenha 2 ou mais valores NaN

```python
df.dropna(thresh= 2)
```

### Removendo dados NaN de columas específicas

```python
df.dropna(subset = ['Year'])
```

# Preenchendo dados nulos

### Utilizando df.fillna()

### Substituindo valores nulos e especificando as colunas

```python
df.fillna(value = {'Year': 'Não Informado', 'Genre': 'Desconhecido'})
```

## Preencher os valores utilizando interpolação ("Replicar o último valor aos dados ausentes")

```python
df.fillna(method = 'ffill')
```

## Preencher valores utilizando a média

```python
df.fillna(df.mean())
```

# Dados Duplicados

```python
df = df.append(pd.Series(), ignore_index=True)
df.tail()
```

## DROPANDO dados duplicados e mantendo a primeira linha

```python
df[df.duplicated()]
```

```python
df.drop_duplicates().tail()
```

## Removendo as linhas duplicadas e MANTENDO  a últiima utilizando keep

```python
# Por padrão o keep vem first
df.drop_duplicates(keep = 'last').tail()
```

### utilizando SUBSET ele considera as colunas que podem ter linhas duplicadas

```python
df.drop_duplicates(subset = ['Name', 'Platform','Year'], keep = 'last').tail()
```

# Transformação de dados com map e lambda

```python
df['Year']
```

```python
df.fillna(value= {'Year': 'Não Informado'}).tail()
```

<h4> Utilizando map lambda para gerar uma nova coluna utilizando outra </h4>

```python
df['Nova_coluna'] = df['Genre'].map(lambda x: str(x)[0:3] if x is not np.NaN else np.NaN)
df
```

<h2> Replace para substituir valores (alternativa do fill) </h2>

```python
df.replace(np.nan, -999)
```

### Utilizando com N valores


```python
df.replace([0.00, 2006.0, np.nan], 'Nao informado')
```

## Passando uma lista como dados substitutos

```python
df.replace([2010.0, 'Sports', np.nan], [2022, 'Não informado', 0])
```

# Discretização de Dados
<p> utilizar <b>pd.cut</b> para discretizar ou separar em compartilhamentos (bins) dados contínuos</p>

```python
bins = [0.0, 10.0, 20.0, 30.0, 50.0, 61.0, 100]
```

```python
vendas = pd.cut(df['Global_Sales'].to_numpy(), bins = bins)
vendas.value_counts()
```

```python
df['Vendas'] = vendas
df['Vendas'].value_counts()
```

```python
df.head()
```

```python
df['Vendas'] = pd.cut(df['Global_Sales'].to_numpy(), bins = bins, 
                      labels = ['Muito-Baixa', 'Baixa', 'Média', 'Alta', 'Muito-Alta', 'Milagre'])
```

```python
df.head()
```

```python
df.Vendas.value_counts(ascending = True)
```
