### About Dataset


<h4> Content</h4>
<h6> This dataset comprises of prices and price % change for coarse wool, copra, cotton, fine wool, harlog, hard sawnwood, hide, plywood, rubber,softlog, sawn wood, and wood pulp

<a><h6> dataset: https://www.kaggle.com/datasets/kianwee/agricultural-raw-material-prices-19902020

##### Tradução livre dos materiais
<li>lã grossa, copra, algodão, lã fina, harlog, madeira serrada dura, couro, compensado, borracha, softlog, madeira serrada e polpa de madeira

```python
# import requerid packages
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv("agricultural_raw_material.csv")
df.head(3)
```

```python
# Checking info about data if has Nun Values
df.isnull().sum()
```

```python
# Replacing '%', ',', '-'
df = df.replace('%', '', regex = True)
df = df.replace(',', '', regex = True)
df = df.replace('-', '', regex = True)
df = df.replace('MAY90', np.nan)
df = df.replace('', np.nan)
```

```python
# Dropping rows with NaN values
df.dropna(inplace = True)
```

```python
# Checking to see if all NaN are resolved
df.isnull().sum()
```

```python
# Collecting the columns is Object (except Month)
lst = df.columns[df.dtypes == 'object'][1::]
```

```python
# Converting data type to float
df[lst] = df[lst].astype('float')
df.dtypes
```

```python
df.head()
```

```python
# Converting the column Month for Year/Month and setting as index
df.Month = pd.to_datetime(df.Month.str.upper(), format = '%b%y', yearfirst= False)
# Indexing Month
df = df.set_index('Month')
```

```python
df.head()
```

<h5> Load the packages 

```python
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

<h4>Heat Map

```python
# Raw materials list
raw_data = ['Coarse wool Price', 'Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price','Hard sawnwood Price',
       'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price','Soft sawnwood Price', 'Wood pulp Price']
# Getting the correlation matrix
corrmat = df[raw_data].corr()

# Setting the size of plot
fig = plt.figure(figsize = (12, 9))

# masking the upper traingle part since matrix is symmetric(repetitive)
mask = np.triu(np.ones_like(corrmat, dtype = bool))
sns.heatmap(corrmat, vmax = .8,mask=mask, square = True, annot = True)
plt.show()
```

<P> O mapa de calor mostra a correlação entre as matérias-primas maior o valor correlacionado maior chance de serem duas matérias-primas relacionadas, mas não necessariamente

<p> Explore uma ou mais colunas traçando um gráfico abaixo e adicione alguma explicação sobre ele matriz de correlação 

```python
plt.figure(figsize=(30,15))
changelist = ['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change',
              'Coarse wool price % Change','Fine wool price % Change','Hard log price % Change',
              'Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']

# Generate a correlation matrix for the whole dataset
corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

<p> A partir deste gráfico podemos dizer que quase não há relação entre a variação percentgual dos preços das matérias-primas
<li> O valor negativo implica que duas variáveis estão negativamente correlacionadas (um aumento, outro descrécimo)
<li> Zero implica nenhuma relação
<li> Caso contrário, maior o valor maior a chance de relação, preços e seus gráficos de % de mudança

<h5> Q1: Descubra a variação normal do preço de cada matéria-prima 

```python
#df[["Wood pulp Price", "Wood pulp price % Change"]].hist(figsize=(11, 9), linewidth=1)

changelist=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change','Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change','Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change']
for i in range(len(changelist)):
    plt.figure(figsize=(12,12))
    df[changelist[i]].hist(figsize=(11, 9), linewidth=1)
    plt.xlabel('% Change')
    plt.ylabel('count')
    plt.legend(changelist[i:],loc='upper center',bbox_to_anchor=(1.2,1))
```
